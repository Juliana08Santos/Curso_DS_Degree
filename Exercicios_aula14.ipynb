```python
#b. Sabendo que as cinco colunas correspondem, nessa ordem, a:

#sepal length (cm)
#sepal width (cm)
#petal length (cm)
#petal width (cm)
#class:
#Iris Setosa
#Iris Versicolor
#Iris Virginica

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
colunas = [ 'sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
result = pd.read_csv(url, names=colunas)
result
```

```python
#c. Crie uma coluna para o volume sabendo que ele é calculado por (pi x petallength x sepal_length^2)/3
result['volume'] = ( np.pi * result['petal_length'] * result['sepal_length']**2 ) / 3
result.head()
```

```python
#c. Calcule a média de cada atributo das flores.
atributos = ['sepal_length', 'sepal_width',	'petal_length', 'petal_width']
result[atributos].mean(axis=0)
```

```python
#d. Novamente, calcule a média de cada atributo das flores só que agora faça-o separadamente para cada um dos tipos de íris indicados na coluna "class". Utilize o loop for para reduzir quantidade de linhas.
#Existe diferença entre as médias de cada um dos atributos das flores?
result.pivot_table(values = atributos, index = 'class', aggfunc = 'mean')
```

```python
#e. Adicione uma nova coluna com o valor da média de 'sepal length' apenas para a íris da classe Iris-setosa.
media_sepal = result[ result['class']=='Iris-setosa']['sepal_length'].mean()

result['media_setosa'] = np.where( result['class'] == 'Iris-setosa', media_sepal, 0)
result
```
