```python
import numpy as np
import pandas as pd
```

```python
#Considere a existência de três tabelas distintas:
#customer.csv : Possui a informação dos clientes em duas colunas: customer id customer name
#products.csv : Contém informação dos produtos vendidos pela empresa em três colunas - p_id (product id), product (name) e price
#sales.csv : Contém informações das vendas realizadas em seis colunas - sale_id, c_id (customer id), p_id (product_id), qty (quantity sold), store (name)
#Conhecendo as bases responda:

#a) Quais foram os produtos vendidos?

dataframe_customer = pd.read_csv("customer.csv", sep=',')
dataframe_products = pd.read_csv("products.csv", sep=',')
dataframe_sales = pd.read_csv("sales.csv", sep=',')
```

```python
vendeu = dataframe_sales['product'].unique()
vendeu
```

```python
#b) E os não vendidos?
produto = dataframe_products['product'].unique()
nao_vendeu = np.isin(produto, vendeu, invert=True)
nao_vendeu = produto[nao_vendeu]
nao_vendeu
```

```python
#c) Quais foram as cinco maiores vendas? Salve essas vendas em um arquivo excel.
vendas = pd.merge(dataframe_sales, dataframe_products, on="p_id")
vendas.drop('product_y', axis=1, inplace=True)
vendas['revenue'] = vendas['qty'] * vendas['price']
vendas.rename(mapper={'product_x':'product'}, axis=1, inplace=True)
vendas_5 = vendas.sort_values(by='revenue', ascending=False).head()

vendas_5.to_excel('vendas_5.xlsx', index = False)
vendas_5
```
